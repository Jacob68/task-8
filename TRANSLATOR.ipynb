```python
import pandas as pd
import nltk
nltk.download("punkt")
nltk.download("wordnet")
from deep_translator import GoogleTranslator
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Jacob68/task-8/main/English.csv")
data.head()
```

```python
Trans_En = pd.DataFrame(data["English words/sentences"].unique()).rename(columns = {0:'English Words'})
print(Trans_En)
```

```python
Trans_En["English Words"] = Trans_En["English Words"].str.replace("[^a-zA-Z0-9]", " ") 
Trans_En["English Words"]
```

```python
Trans_En["English Words"].head(1000)
```

WE ARE ONLY TAKIN 150 SAMPLES OUT OF ALL DATA SO HEAD=150

ENGLISH TO SPANISH

```python
def Translate_Es(number):
    translation_tokenized=GoogleTranslator(source='english', target='es').translate(number)
    return translation_tokenized

Trans_Spanish_Words = [Translate_Es(r) for r in Trans_En["English Words"].head(150)]
Trans_Spanish_Words
```

```python
Trans_Spanish_Sentence = [Translate_Es(r) for r in Trans_En["English Words"].tail(150)]
Trans_Spanish_Sentence
```

ENGLISH TO FRENCH

```python
def Translate(rev):
    translation_tokenized = GoogleTranslator(source='english', target='fr').translate(rev)
    return translation_tokenized
Trans_French_Words = [Translate(r) for r in Trans_En["English Words"].head(150)]
Trans_French_Words
```

```python
Trans_French_Sentence = [Translate(r) for r in Trans_En["English Words"].tail(150)]
Trans_French_Sentence
```

ENGLISH TO GERMAN

```python
def Translate_De(rev):
    translation_tokenized = GoogleTranslator(source='english', target='de').translate(rev)
    return translation_tokenized
Trans_German_Words = [Translate_De(r) for r in Trans_En["English Words"].head(150)]
Trans_German_Words
```

```python
Trans_German_Sentence = [Translate_De(r) for r in Trans_En["English Words"].tail(150)]
Trans_German_Sentence
```
